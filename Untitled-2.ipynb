```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
x = pd.read_csv('/Users/krishisrani/Downloads/titanic/train.csv')
```

```python
from sklearn.model_selection import train_test_split

train, test = train_test_split(x, test_size=0.2)
```

```python
correlation = train[['Survived','Pclass']]
correlation
```

```python
grouped = correlation.groupby(['Pclass', 'Survived']).size().reset_index()
grouped
```

```python
mask = (grouped['Survived'] == 0)
No_survival = grouped.loc[mask]
No_survival.reset_index(drop=True)
```

```python
No_survival = No_survival.drop('Survived', axis=1)
No_survival.reset_index(drop=True)
```

```python
mask = (grouped['Survived'] == 1)
Survived = grouped.loc[mask]
Survived.reset_index(drop=True)
```

```python
Survived = Survived.drop('Survived', axis=1)
Survived.reset_index(drop=True)
```

```python
plot = No_survival.plot.pie(y=0, figsize=(5, 5), ylabel='', labels = No_survival['Pclass'], title = "Did not survive")
```

```python
survived_plot = Survived.plot.pie(y=0, figsize=(5, 5), ylabel='', labels = Survived['Pclass'], title = "Survived")
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(solver: newton-cg)
lr.fit(correlation['Pclass'].values.reshape(-1, 1),train['Survived'] )
```

```python
Actual_survived = test[['Survived']].reset_index(drop=True)
Actual_survived
```

```python
test1 = test.drop(columns= ['Survived'])
test1.reset_index(drop=True)
```

```python
testing1 = lr.predict(test1['Pclass'].values.reshape(-1, 1))
```

```python
testing1
```

```python
from sklearn import metrics

```

```python
metrics.f1_score(y_true = Actual_survived, y_pred = testing1)
```
